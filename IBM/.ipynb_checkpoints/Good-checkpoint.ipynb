```python
data = pd.read_csv('Submissions/titanic_train.csv',index_col = 0)
```

```python
data['Sex'] = data['Sex'].map({'male': 0 , 'female' : 1 })
data['Embarked'] = data['Embarked'].map({'S' : 0, 'C' : 1, 'Q' : 2})
```
