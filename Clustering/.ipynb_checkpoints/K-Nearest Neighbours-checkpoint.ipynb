```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
import operator
```

```python
class Dataset(object) :
    def __init__(self,filen) :
        self.filename = filen
        self.data = pd.read_csv(self.filename)
        self.datam = self.data.as_matrix()
        self.x = self.datam[:,[1]]
        self.y = self.datam[:,[2]]
```

```python
datas = Dataset("AirPassengers.csv")
```

```python
''' PICTURE OF THE DATASET '''
%matplotlib inline
plt.plot(datas.x , datas.y)
plt.scatter(datas.x,datas.y)
plt.show()
```
